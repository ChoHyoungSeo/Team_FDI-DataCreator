```python
from __future__ import absolute_import, division, print_function, unicode_literals, unicode_literals

from sklearn.preprocessing import OneHotEncoder
# tensorflow와 tf.keras를 임포트합니다
import tensorflow as tf
from tensorflow import keras

import torch 

# 헬퍼(helper) 라이브러리를 임포트합니다
import numpy as np
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
print(tf.__version__)
print(torch.__version__)
```

```python
X_train = pd.read_csv('sample_data/X_train.csv', index_col=0)
Y_train = pd.read_csv('sample_data/Y_train.csv')
X_test = pd.read_csv('sample_data/X_test.csv')
Y_test = pd.read_csv('sample_data/Y_test.csv')

X_train = X_train.drop('td', axis='columns')
X_train=X_train.drop('code', axis='columns')
X_train=X_train.drop('X', axis='columns')
X_train=X_train.drop(X_train.index[[1,3]])
X_train=X_train.iloc[2:]

X_test = X_test.drop('td', axis='columns')
X_test=X_test.drop('code', axis='columns')
X_test=X_test.drop('X', axis='columns')
X_test=X_test.drop(X_test.index[[1,3]])
X_test=X_test.iloc[2:]
X_test = X_test.iloc[4:]
X_test=X_test.drop(X_test.columns[0], axis=1)

Y_train = Y_train.drop('td', axis='columns')
Y_train=Y_train.drop('code', axis='columns')
Y_train = Y_train.iloc[4:]
Y_train=Y_train.drop(Y_train.columns[0], axis=1)

Y_test = Y_test.drop('td', axis='columns')
Y_test=Y_test.drop('code', axis='columns')
# Y_train=Y_train.drop('X', axis='columns')
Y_test=Y_test.drop(Y_test.index[[1,3]])
Y_test=Y_test.iloc[2:]
Y_test = Y_test.iloc[4:]
Y_test=Y_test.drop(Y_test.columns[0], axis=1)

X_train = X_train.values
Y_train = Y_train.values
X_test = X_test.values
Y_test = Y_test.values
```

```python
model = keras.Sequential([
    keras.layers.Dense(1500, activation='relu'),
    keras.layers.Dense(1000, activation='relu'),
    keras.layers.Dense(800, activation='relu'),
    keras.layers.Dense(600, activation='relu'),
    keras.layers.Dense(300, activation='relu'),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(5, activation='softmax')
])
```

```python
model.compile(optimizer='SGD',
             loss = 'sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, epochs=2)
```

```python
test_loss, test_acc = model.evaluate(X_test, Y_test, verbose=2)
print("\n 테스트 정확도: ", test_acc)
```
